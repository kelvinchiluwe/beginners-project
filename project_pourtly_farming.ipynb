```python
df=pd.read_csv('flock_average_weight_chicken_data.csv')
```

```python
def Weekly_Feed_Cost():
    while True:

            X=df[['Average Weight (kg)','Age (weeks)','Number of Chickens','Breed_Broiler','Breed_Hybrid','Breed_Layer']].values
            y=df['Weekly Feed Cost ($)'].values
            # X=(X-X.mean())/X.std()
            X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
            pourtly=LinearRegression()
            pourtly.fit(X_train,y_train,)
            try:
                # Give a farmer an access to what he want to use depending on the following
                print('====enter the following factors====')
                avg_weight = float(input("Enter Average Weight (kg): "))
                age_weeks = int(input("Enter Age (weeks): "))
                num_chickens = int(input("Enter Number of Chickens: "))
                breed_broiler = int(input("Is Breed Broiler? (1 for Yes, 0 for No): "))
                breed_hybrid = int(input("Is Breed Hybrid? (1 for Yes, 0 for No): "))
                breed_layer = int(input("Is Breed Layer? (1 for Yes, 0 for No): "))
                user_input = np.array([[avg_weight, age_weeks, num_chickens, breed_broiler, breed_hybrid, breed_layer]])
                prediction=pourtly.predict(user_input)
                print(f'your weekly feed cost is ${prediction} depending on your information you gave me')
                break
            except Exception as e:
                print('please enter a valid number....next_time>>>>>>>')
                # break            
```

```python
def Weekly_Feed_Consumption():
    while True:
        X=df[['Average Weight (kg)','Age (weeks)','Number of Chickens','Breed_Broiler','Breed_Hybrid','Breed_Layer']].values
        y=df["Weekly Feed Consumption (kg)"].values
        X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
        pourtly=LinearRegression()
        pourtly.fit(X_train,y_train)
        try:
            print('====enter the following factors====')
            avg_weight = float(input("Enter Average Weight (kg): "))
            age_weeks = int(input("Enter Age (weeks): "))
            num_chickens = int(input("Enter Number of Chickens: "))
            breed_broiler = int(input("Is Breed Broiler? (1 for Yes, 0 for No): "))
            breed_hybrid = int(input("Is Breed Hybrid? (1 for Yes, 0 for No): "))
            breed_layer = int(input("Is Breed Layer? (1 for Yes, 0 for No): "))
            user_input = np.array([[avg_weight, age_weeks, num_chickens, breed_broiler, breed_hybrid, breed_layer]])
            prediction=pourtly.predict(user_input)
            print(f'your weekly feed consumption is {prediction}kg depending on your information you gave me')
            break
        except exception as e:
            print('enter a number next time')       
```

```python
def Weekly_Vaccination_Doses():
    while True:
        X=df[['Average Weight (kg)','Age (weeks)','Number of Chickens','Breed_Broiler','Breed_Hybrid','Breed_Layer','Disease_Fowl Pox','Disease_Gumboro','Disease_Infectious Bronchitis','Disease_Newcastle Disease']].values
        y=df['Weekly Vaccination Doses'].values
        X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
        pourtly=LinearRegression()
        pourtly.fit(X_train,y_train)
        try:
            print('====enter the following factors====')
            avg_weight = float(input("Enter Average Weight (kg): "))
            age_weeks = int(input("Enter Age (weeks): "))
            num_chickens = int(input("Enter Number of Chickens: "))
            breed_broiler = int(input("Is Breed Broiler? (1 for Yes, 0 for No): "))
            breed_hybrid = int(input("Is Breed Hybrid? (1 for Yes, 0 for No): "))
            breed_layer = int(input("Is Breed Layer? (1 for Yes, 0 for No): "))
            disease_fowl_pox = int(input("Has Disease Fowl Pox? (1 for Yes, 0 for No): "))
            disease_gumboro = int(input("Has Disease Gumboro? (1 for Yes, 0 for No): "))
            disease_ib = int(input("Has Infectious Bronchitis? (1 for Yes, 0 for No): "))
            disease_nd = int(input("Has Newcastle Disease? (1 for Yes, 0 for No): "))
        

          # Format user input for prediction
            user_input = np.array([[avg_weight, age_weeks, num_chickens, breed_broiler, breed_hybrid, breed_layer,disease_fowl_pox, disease_gumboro, disease_ib, disease_nd]])
            prediction=pourtly.predict(user_input)
            print(f'your Weekly_Vaccination_Doses is {prediction}doses depending on your information you gave me')
            break
        except Exception as e:
            print('next time enter numbers only............')
```

```python
def menu():
    while True:
        print('/n<<<< welcome to chicken farming digital system>>>>')
        try:
            print('1.Weekly Feed Cost ($)')
            print('2.Weekly Feed Consumption')
            print('3.Weekly Vaccination Doses')
            print('4.exit....program..')
            option=int(input('enter your option:   '))
            if option==1:
                Weekly_Feed_Cost()
            elif option==2:
                Weekly_Feed_Consumption()
            elif option==3:
                Weekly_Vaccination_Doses()
            elif option==4:
                print('exiting the program.........')
                break
            choice=input('do you want to continue?(yes or no):  ')
            if choice=='yes':
                print('lets roll... again>>>>')
                continue
            elif choice=='no':
                print('exiting the program..........')
                # quit()
                break
        except Exception as e:
            print('invalid input!!!!!!!!!')        
```

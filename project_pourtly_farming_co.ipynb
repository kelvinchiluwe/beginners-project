```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

def Weekly_Vaccination_Doses_():
    while True:
        # Selecting relevant features and target variable
        X = df[['Average Weight (kg)', 'Age (weeks)', 'Number of Chickens', 
                'Breed_Broiler', 'Disease_Fowl Pox', 'Disease_Gumboro', 
                'Disease_Infectious Bronchitis', 'Disease_Newcastle Disease']].values
        y = df['Weekly Vaccination Doses'].values
        
        # Splitting data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.0001, random_state=42)

        # Standardizing the features
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)

        # Initialize Random Forest model
        poultry_model = RandomForestRegressor(n_estimators=1000,min_samples_split=7, random_state=42)
        poultry_model.fit(X_train, y_train)

        try:
            # Collect user input
            print('====Enter the following factors====')
            avg_weight = float(input("Enter Average Weight (kg): "))
            age_weeks = int(input("Enter Age (weeks): "))
            num_chickens = int(input("Enter Number of Chickens: "))
            breed_broiler = int(input("Is Breed Broiler? (1 for Yes, 0 for No): "))
            disease_fowl_pox = int(input("Has Disease Fowl Pox? (1 for Yes, 0 for No): "))
            disease_gumboro = int(input("Has Disease Gumboro? (1 for Yes, 0 for No): "))
            disease_ib = int(input("Has Infectious Bronchitis? (1 for Yes, 0 for No): "))
            disease_nd = int(input("Has Newcastle Disease? (1 for Yes, 0 for No): "))

            # Format user input for prediction
            user_input = np.array([[avg_weight, age_weeks, num_chickens, breed_broiler, 
                                    disease_fowl_pox, disease_gumboro, disease_ib, disease_nd]])
            user_input_normal = scaler.transform(user_input)
            
            # Make prediction
            prediction = poultry_model.predict(user_input_normal)
            print(f'Your Weekly Vaccination Doses estimate is {prediction[0]:.2f} doses.')

            break
        except Exception as e:
            print('Please enter valid numbers only...')
Weekly_Vaccination_Doses()
```
